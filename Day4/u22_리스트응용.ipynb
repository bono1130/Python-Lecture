```python
# 소수(prime number) - 2, 3, 5, 7, ...
prime_numbers = []

def is_prime_number(num):
    if num == 2:
        return True
    for i in range(2,num):
        if num % i == 0:
            return False
    return True

for i in range(2,101):
    if is_prime_number(i):
        prime_numbers.append(i)
print(prime_numbers)
```

```python
# 1에서 100까지 7의 배수 리스트 만들기
multiple7 = []
for i in range(1,101):
    if i % 7 == 0:
        multiple7.append(i)
multiple7
```

- 정렬 (Sort)

```python
a = [10,20,50,30,40]
a.sort()
a
```

```python
# 반복문으로 sort 구현
a = [10, 20, 50, 30, 40, 5, 25]
for i in range(len(a)-1):
    for k in range(i+1,len(a)):
        if a[i] > a[k]:
            a[i], a[k] = a[k], a[i]
a
```

```python
# 올림차순(ascending order)
a = [10, 20, 50, 30, 40, 5, 25]
a.sort()
a
```

```python
# 내림차순(descending order)
a = [10, 20, 50, 30, 40, 5, 25]
a.sort(reverse = True)
a
```

### 22.5 리스트 표현식(comprehension)

```python
# 1에서 10까지 제곱수로 된 리스트 만들기
sq_list = []
for i in range(1,11):
    sq_list.append(i*i)
sq_list
```

```python
sq_list = [i*i for i in range(1,11)]
sq_list
```

```python
odd_sq_list = []
for i in range(1,11):
    if i % 2 ==1:
        odd_sq_list.append(i*i)
odd_sq_list
```

```python
odd_sq_list = [i*i for i in range(1,11) if i % 2 == 1]
odd_sq_list
```

```python
# 구구단
a = [i*k for i in range(2,10) for k in range(1,10)]
print(a)
```

```python
a = [i*k for i in range(2,10) 
            for k in range(1,10)]
print(a)
```

### 22.6 map 적용

```python
a = [1.2, 2.5, 3.7, 4.6]
b = []
for element in a:
    b.append(int(element))
b
```

```python
b = list(map(int, a))
b
```

```python
b = [int(i) for i in a]
b
```

```python
# a의 제곱수로 된 리스트
sq_a = []
for element in a:
    sq_a.append(element*element)
sq_a
```

```python
sq_a = [element*element for element in a]
sq_a
```

```python
# map 적용
def square(x):
    return x*x
sq_a = list(map(square, a))
sq_a
```

```python
# 익명(lambda) 함수와 map 적용
sq_a = list(map(lambda x: x*x, a))
sq_a
```
