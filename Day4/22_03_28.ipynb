- 22.10 심사문제: 2의 거듭제곱 리스트 생성하기

```python
start, stop = map(int, input().split())
sq_2 = [2 ** i for i in range(start, stop + 1)]
sq_2.pop(1)
sq_2.pop(-2)
print(sq_2)
```

- 23.7 심사문제: 지뢰찾기

```python
import numpy as np
from pprint import pprint
matrix_col, matrix_row = map(int, input().split())
matrix = []
for i in range(matrix_row):
    matrix.append(list(input()))
matrix_padding = np.pad(matrix, ((1,1),(1,1)), 'constant', constant_values='.')
pprint(matrix_padding,indent=1, width=20)  
```

```python
for i in range(1,matrix_row+1):
    for k in range(1,matrix_col+1):
        count = 0
        if matrix_padding[i][k] == '.':
            for a in range(i-1,i+2):
                for b in range(k-1,k+2):
                    if matrix_padding[a][b] == '*':
                        count += 1
            matrix_padding[i][k] = count
            print(matrix_padding[i][k], end ='')
        else:
            print(matrix_padding[i][k],end='')
    print()
```

연습문제
<pre>
a + b + c = 1000, a < b < c 를 만족하고,
a*a + b*b = c*c를 만족하는 파타고라스 수 a, b, c는?
</pre>

```python
#a + b + c = 1000
for a in range(1,1000):
    for b in range(a+1,1000-a):
        c = 1000 - a - b
        if a**2 + b**2 == c**2:
            print(a,b,c)
```
