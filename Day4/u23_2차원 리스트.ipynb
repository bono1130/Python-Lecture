### 23.2 반복문과 2차원 리스트

```python
a = [[10,20],[30,40],[50,60]]
for row in a:
    print(row)
```

```python
for x,y in a:
    print(x,y)
```

```python
for row in a:
    for element in row:
        print(element,end=' ')
    print()
```

```python
a = [[10,20],
     [500,600,700],
     [9],
     [30,40],
     [8],
     [800,900,1000]]
for row in a:
    for element in row:
        print(element,end=' ')
    print()
```

```python
len(a), len(a[0]),len(a[1])
```

```python
for i in range(len(a)):
    for k in range(len(a[i])):
        print(a[i][k], end =' ')
    print()
```

### 23.3 반복문으로 리스트 만들기

```python
# [[10,11],[20,21],[30,31]]
for i in range(3):
    for k in range(2):
        print(10*(i+1)+k, end=' ')
    print()
```

```python
outer_list = []
for i in range(3):
    inner_list = []
    for k in range(2):
        inner_list.append(10*(i+1)+k)
    outer_list.append(inner_list)
pprint(outer_list, indent = 1, width =20)
```

```python
a = []
for i in range(3):
    for k in range(2):
        a.append(10*(i+1)+k)
a
```

```python
two_d = []
for i in range(3):
    line = []
    for k in range(2):
        line.append(10*(i+1)+k)
    two_d.append(line)
two_d
```

- 리스트 표현식

```python
# 바깥쪽 : [... for i in range(3)]
# 안쪽 : [10*(i+1)+k for k in range(2)]
[[10*(i+1)+k for k in range(2)] for i in range(3)]
```

```python
# [[1,2,3,4],[11,12,13,14],[21,22,23,24]]
# 행 인덱스(i): 0,1,2 - range(3)
# 열 인덱스(k): 0,1,2,3 - range(4)
# 일반항 : 10 * i + k +1
[[10*i+k+1 for k in range(4)] for i in range(3)]
```

```python
test1 = []
for i in range(3):
    line = []
    for k in range(4):
        line.append(10*i+k+1)
    test1.append(line)
test1
```
