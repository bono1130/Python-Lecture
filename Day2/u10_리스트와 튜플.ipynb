- range 객체

```python
# range parameter가 1개 : 횟수를 의미 (0 ~ n-1)
list(range(5))
```

```python
# range parameter가 2개 : 시작과 끝을 의미 (시작 ~ 끝-1)
list(range(1,10))
```

```python
# range parameter가 3개 : 시작, 끝, 증가폭을 의미
list(range(1,10,2))
```

```python
# 음수의 경우 시작에서부터 폭이 줄어듬 / 마지막은 1 큰 수에서 끝남
list(range(10,0,-2))
```

### 10.2 튜플(tuple)

```python
a = (38, 21, 53, 62, 19)
```

```python
person = ('james', 27, 175.3, '남')
person
```

```python
size = (3)
type(size)
```

```python
# element가 1개인 튜플은 ,를 붙여줘야함
size = (3,)
type(size)
```

```python
size = 3,
type(size)
```

- 튜플은 요소를 변경, 추가, 삭제할 수 없다.

```python
t = ()
type(t)
```

```python
# 최초 한 번 반복문으로 만들 수는 있으나, 변경이 안됨
t = tuple(range(5))
t
```

- 리스트 <-> 튜플

```python
l = list(t)
type(l)
```

```python
type(tuple(l))
```

- list/tuple unpacking

```python
a, b, c, d = 'a', 'b', 'c', 'd'
a, b, c, d
```

```python
a, b, c, d = list('abcd')
a, b, c, d
```

```python
x, y, z = range(3)
x, y, z
```

```python
x, y, z = input().split()
x, y, z
```

```python
x, y, z = map(int, input().split())
x, y, z
```

```python
x, y, z = map(lambda x : int(x) * 2 + 5, input().split())
x, y, z
```
