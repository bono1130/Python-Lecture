# Unit 16. 반복문

### 16.1 for, range

```python
for i in range(3):
    print('Hello World!')
```

```python
for i in range(3):
    print('안녕하세요.', i)
```

### 16.2 for, range 응용

```python
for i in range(5, 12):
    print(i, end = ' ')
```

```python
for i in range(10, 0, -1):
    print(i, end = ' ')
```

```python
# Loop 안에서 index값이 필요하지 않으면 i 대신 _를 쓸 수 있다.
for _ in range(3):
    print('Hello')
```

```python
# for-loop에서는 loop 안에서 index 값을 변경시키지 말 것
for i in range(10):
    print(i, end = ' ')
    i +=5           # 금기 사항
```

```python
count = int(input())
for i in range(count):      # count가 0이면 반복문이 실행되지 않음
    print(i, end=' ')
```

### 16.3 시퀀스 객체로 반복하기

```python
a = [10, 20, 30, 40, 50]
for element in a:
    print(element, end=' ')
```

```python
fruits = ('사과', '배', '감', '귤')
for fruit in fruits:
    print(fruit)
```

```python
for letter in 'python':
    print(letter, end=' ')
```

```python
for letter in reversed('python'):
    print(letter, end=' ')
```

```python
for letter in 'python'[::-1]:
    print(letter, end=' ')
```

- 반복문 응용

```python
# 1에서 100까지 합
sum = 0
for i in range(1, 101):
    sum += i
print(sum)
```

```python
# 1에서 10까지 sqrt 값을 갖는 리스트를 만드세요.
import math
sqrt = []
for i in range(1, 11):
    sqrt.append(round(math.sqrt(i),4))
print(sqrt)
```
