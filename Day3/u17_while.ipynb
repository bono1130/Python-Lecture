# Unit 17. while 반복문

### 17.1 while 반복문

```python
i = 0
while i < 5:
    print('Hello World',i)
    i += 1
```

### 17.2 반복 횟수가 정해져 있지 않은 경우

```python
import random
```

```python
dice = 0
while dice != 3:
    dice = random.randint(1, 6)
    print(dice, end = ' ')
```

```python
while True:                             # 무한 루프
    dice = random.randint(1, 6)
    print(dice, end = ' ')
    if dice == 3:
        break
```
