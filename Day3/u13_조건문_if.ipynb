### 13.2 if 조건문과 들여쓰기

```python
if x == 10:
    print(x, '입니다.')
    print('들여쓰기 한 이 부분은 조건이 참일 경우에 실행됩니다.')
print('들여쓰기가 끝나면 조건에 관계없이 실행됩니다.')
```

### 13.3 중첩(Nested) 조건문

```python
x = random.randint(1,20)
if x >= 10:
    print('10 이상입니다.')
    if x>= 15:
        print('15 이상입니다.')
print(x)
```

```python
# 10 이상 정수, 3의 배수를 찾기

x = random.randint(1,20)
if x >= 10:
    print('10 이상입니다.')
    if x % 3 == 0:
        print('3의 배수입니다.')
print(x)
```
