# Unit 19. 중첩(Nested) 반복문

### 19.1 중첩 loop

```python
for i in range(1, 6):
    for k in range(i):
        print('*', end = '')
    print()
```

```python
for i in range(1, 6):
    print('*' * i)
```

```python
width, height = 5, 3
for i in range(height):
    for k in range(width):
        print('*',end ='')
    print()
```

```python
for i in range(1,6):
    for k in range(5-i):
        print(' ', end = '')
    for j in range(i):
        print('*', end = '')
    print()
```

```python
# 가운데 행의 별의 개수를 입력으로 받아서 다이아몬드 그리기
num = 11
n = num // 2 + 1
for i in range(n):
    for k in range(n - 1 - i):
        print(' ', end = '')
    for k in range(i*2+1):
        print('*', end = '')
    print()
for i in range(n-2, -1, -1):
    for k in range(n - 1 - i):
        print(' ', end = '')
    for k in range(2 * i + 1):
        print('*', end = '')
    print()
```
