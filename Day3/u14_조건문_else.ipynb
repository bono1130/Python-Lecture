# Unit 14. 조건문 if - else

### 14.1 else

```python
import random
```

```python
x = random.randint(1,6)
if x >= 4:
    print('큰 수')
else:
    print('작은 수')
print(x)
```

```python
x = random.randint(1,6)
if x % 2 == 0:
    print('짝수')
else:
    print('홀수')
print(x)
```

```python
x = random.randint(1,6)
if x % 2 == 0:
    result = 'even number'
else:
    result = 'odd number'
print(x, result)
```

- 조건부 표현식 (3항 연산자)

```python
# 암기하기
# Java, JS에서는 result = x % 2 == 0 ? "even" : "odd";
result = 'even' if x % 2 == 0 else 'odd'
print(x, result)
```

### 14.2 else와 들여쓰기

### 14.3 동작 방식

```python
x = ''
if x:
    print(x)
else:
    print('0, 0.0, "", False, None 입니다')
```

### 14.4 여러 개의 조건식

```python
# 3과 5의 공배수면 FizzBuzz를 출력
# 위의 경우가 아니면 숫자값을 출력

x = 30
if x % 3 == 0 and x % 5 == 0:
    print('FizzBuzz')
else:
    print(x)
```

```python
x = 14
if x >= 10 and x <= 20:
    print('10이상이고 20이하')
else:
    print(x)
```

```python
x = 14
if 10 <= x <= 20:
    print('10이상이고 20이하')
else:
    print(x)
```

```python
year = int(input("enter a year: "))
 
if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
    print("윤년")
else:
    print("평년")
```
