# Unit 20. FizzBuzz

```python
for i in range(1, 101):
    if i % 15 == 0:
        print('FizzBuzz', end = ' ')
    elif i % 3 == 0:
        print('Fizz', end = ' ')
    elif i % 5 == 0:
        print('Buzz', end = ' ')
    else :
        print(i, end = ' ')
```

```python
for i in range(1, 101):
    print('Fizz' * (i % 3 == 0) + 'Buzz' * (i % 5 == 0) or i, end = ' ')
```

### 연습문제
<pre>
세자리수 두 개를 곱해서 만든 수가 대칭수(palindrome)일 때
가장 큰 대칭수와 그 경우 두 가지를 구하시오
</ore>

```python
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

max = 0

for i in range(100, 1000):
    for k in range(i, 1000):
        mul = i * k
        if is_palindrome(mul):
            if max < mul:
                max = mul
                num1, num2 = i, k
print(num1, num2, max)
```
