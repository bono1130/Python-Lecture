# Unit 15. 조건문 elif

```python
import random
```

```python
# 90점 이상 A, 80점 이상 B, 70점 이상 C, 60점 이상 D, 나머지 F
score = random.randint(40, 100)
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
print(score, grade)
```

```python
score = random.randint(40, 100)
grade = 'A' if score >= 90 else\
            'B' if score >= 80 else\
                'C' if score >= 70 else\
                    'D' if score >= 60 else 'F'
print(score, grade)
```

### 연습문제
<pre>
생일을 입력으로 받아 오늘 현재의 만 나이를 게산하세요.
</pre>

```python
import datetime as dt
today = dt.date.today()
today.year, today.month, today.day
print('생일을 입력하세요(year month day)')
year, month, day = map(int, input().split())
age = today.year - year
if month > today.month or (month == today.month and day > today.day):
    age -= 1
print(age)
```
