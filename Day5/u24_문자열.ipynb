- 문자열 분리(split)

```python
s = 'apple pear grape pineapple orange'
d = s.split()
d
```

```python
s = 'apple,pear,grape,pineapple,orange'
d = s.split(',')
d
```

```python
stopwords = ['a', 'the', 'I', 'am', 'is']
s = 'I am a graduate student.'
d = s.split()
d = [word for word in d if word not in stopwords]
d
```

```python
'apple,pear,grape,pineapple,orange'.replace(',',' ')
```

```python
s = 'I am a graduate student.'
' '.join(word for word in d if word not in stopwords)
```

```python
# 풀어서 쓰기
d = s.split()
sentence = []
for word in d:
    if word not in stopwords:
        sentence.append(word)
sentence = ' '.join(sentence)
sentence
```

### 연습문제
<pre>
1. 디지털 시계(00:00 ~23:59)가 있다.
하룻동안 3이라는 숫자가 표시되는 시간은 몇초인가?

2. 1에서 1000까지 숫자에서 0, 1,...,9는 몇 번 나오나>
</pre>

```python
time_m = 0
for i in range(24):
    if '3' in str(i):
        time_m += 60
    else:
        for k in range(60):
            if '3' in str(k):
                time_m += 1
time = time_m*60
time
```

```python
num_string = ''
for i in range(1,1001):
    num_string += str(i)
```

```python
for i in range(10):
    count = num_string.count(str(i))
    print(i, count)
```

### 24.2 문자열 서식 지정자와 포맷팅

```python
s = 'python'        # 문자열
f'==={s:10s}==='
```

```python
d = 3               # 정수
f'==={d:4d}:{d:02d}==='
```

```python
year, month, day, hour, minute = 2022, 3, 29, 11, 36
f'오늘은 {year}-{month:02d}-{day:02d}이고, 시간은 {hour:02d}:{minute:02d}입니다.'
```

```python
price1, price2 = 35000, 1200000
print(f'가격1: {price1:10,d}원')
print(f'가격2: {price2:10,d}원')
```

```python
import math
f'파이값은 {math.pi:8.5f} 입니다.'
```
