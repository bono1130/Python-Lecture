### 29.2 덧셈함수

```python
# callee, a와 b의 값은 함수가 호출되어지는 순간 결정이 됨
def add(a, b):
    """숫자 두개를 입력하면 덧셈 결과를 출력해주는 함수
    이것을 독스트링(Docstring)이라고 한다."""
    print(a+b)
```

```python
# caller
add(3,4)
```

```python
add(10,20)
```

### 29.3 함수결과 반환

```python
a, b =12000, 24000
f'{a} + {b} --> {a+b}'
```

```python
f'{a} + {b} --> {add(a,b)}'
```

```python
f'{a:,d} +{b:,d} --> {a+b:,d}'
```

```python
#결과값을 반환해주는 함수
def add2(a,b):
    return a+b
```

```python
f'{a:,d} +{b:,d} --> {add2(a,b):,d}'
```

```python
x = add2(10,20)
x
```

```python
def even_odd(num):
    if num %2 == 0:
        return 'Even'
    else:
        return 'Odd'
```

```python
def even_odd2(num):
    if num %2 == 0:
        return 'Even'
    return 'Odd'
```

```python
even_odd(9), even_odd2(9)
```

### 연습문제
<pre>
약수를 구한 뒤 리스트로 만들어서 반환하는 함수 get_divisors(num)을 만드세요.
</pre>

```python
def get_divisors(num):
    divisors = [i for i in range(1,num+1) if num%i == 0]
    return divisors
```

```python
get_divisors(10)
```
