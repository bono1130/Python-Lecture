### Unit 31. 재귀 호출(Recursive call)

- factorial
<pre>
n * (n-1)!        n > 0,
0                 n = 0

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
```

```python
fact(3)
```

```python
fact(5)
```

- fibonacci 수열
<pre>
fibo(n) = fibo(n-1) + fibo(n-2),     n >= 2
fibo(1), fibo(0) = 1,                n = 0,1

```python
def fibo(n):
    if n == 1 or n == 0:
        return 1
    return fibo(n-1) + fibo(n-2)
```

```python
for i in range(11):
    print(f'{i:2d}: {fibo(i)}')
```
