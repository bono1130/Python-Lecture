### 30.1 위치 인수

```python
print(10,20,30,40,50)
```

```python
def print_numbers(a,b,c):
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10,20,30)
```

```python
x = [10,20,30]
print_numbers(*x)           # 리스트 언패킹
```

- 가변 인수

```python
def print_numbers(*args):
    for arg in args:
        print(arg)
```

```python
print_numbers(10,20)
```

```python
print_numbers(10,20,30,40)
```

```python
print_numbers()
```

### 30.2 키워드 인수

```python
def personal_info(name, age, address):
    print('이름:', name)
    print('나이:', age)
    print('주소:', address)
```

```python
# 위치 인수로 호출
personal_info('홍길동', 25, '서울시 서초구 서초동')
```

```python
# 위치 인수로 호출할 경우에 정확히 위치를 지정해야 한다
personal_info(25, '홍길동', '서울시 서초구 서초동')
```

```python
# 키워드로 호출
personal_info(age = 25, name = '홍길동', address = '서울시 서초구 서초동')
```

### 30.3 딕셔너리 언패킹

```python
x = {'name':'홍길동', 'age':30, 'address':'서울시 서초구 서초동'}
personal_info(**x)
```

- 가변 키워드 인수

```python
def personal_info2(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}:{value}')
```

```python
personal_info2(**x)
```

```python
x['성별'] = '남자'
x['고향'] = '서울'
x
```

```python
personal_info2(**x)
```

- 호출하는 순서: 위치 인수가 키워드 인수보다 먼저 나와야 함

```python
personal_info('홍길동', address ='경기도 용인시 수지구', age = 28)
```

```python
personal_info(address ='경기도 용인시 수지구', '홍길동', age = 28)
```

- 가변 위치 인수, 가변 키워드 인수

```python
def custom_print(*args, **kwargs):
    print(*args, **kwargs)
```

```python
custom_print(10, 20, 30, sep=', ', end='')
```

### 30.4 매개변수에 초기값 지정
