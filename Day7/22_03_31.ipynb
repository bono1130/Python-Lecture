- 34.6 심사문제: 게임 캐릭터 클래스 만들기

```python
class Annie:
    def __init__(self, health, mana, ability_power):
        self.health = health
        self.mana = mana
        self.ability_power = ability_power

    def tibbers(self):
        print(f'티버: 피해량 {self.ability_power * 0.65 + 400}')

health, mana, ability_power = map(float, input().split())
 
x = Annie(health=health, mana=mana, ability_power=ability_power)
x.tibbers()
```

2. Account 클래스 만들기
<pre>
ano : 문자열 6글자, '123456'
owner : 문자열
balance: 잔액, 비공개 속성
Method
생성자
deposit(amount) 잔액이 1000만원 이상이면 입금할 수 없음
withdraw(amount) 잔액이 0원 미만이면 출금할 수 없음
출력 가능하게(__str__)
</pre>

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def add(self):
        return self.num1 + self.num2
    def subtract(self):
        return self.num1 - self.num2
    def multiply(self):
        return self.num1 * self.num2
    def divide(self):
        return self.num1 / self.num2

x = int(input("첫 번째 수를 입력하세요. :"))
y = int(input("두 번째 수를 입력하세요. :"))

a = Calculator(x,y)
if x == 0 or y == 0:
    print('0은 계산할 수 없습니다.')
else:
    print(f"{x} + {y} = {a.add()}")
    print(f"{x} - {y} = {a.subtract()}")
    print(f"{x} * {y} = {a.multiply()}")
    print(f"{x} / {y} = {a.divide()}")
```

```python
class Account:
    def __init__(self, owner, ano, balance):
        self.owner = owner
        self.ano = ano
        self.__balance = balance

    def deposit(self, amount):
        if(self.__balance >= 1000000):
            return 1
        else:
            self.__balance += amount
    
    def withdraw(self, amount):
        if(self.__balance < amount):
            return 1
        else:
            self.__balance -= amount
            return 0

    def print_accountInfo(self):
        return "{:>10} {:>10} {:>10}".format(self.owner, self.ano, self.__balance)
```
